```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.linear_model

dataframe = pd.read_csv('dyntime.csv')

fig = plt.figure()
axes = plt.axes(projection='3d')
axes.scatter(dataframe['dbsize'], dataframe['query_size'], dataframe['dynamic_time'])
axes.set_xlabel('dbsize')
axes.set_ylabel('qsize')
axes.set_zlabel('time', rotation=90)
axes.zaxis.labelpad = -1

model = sklearn.linear_model.LinearRegression()
model.fit(dataframe[['dbsize', 'query_size']], dataframe['dynamic_time'])
predictions = model.predict(dataframe[['dbsize', 'query_size']])
print("Equation: y = {:.2f} + {:.2f}x1 + {:.2f}x2".format(model.intercept_, model.coef_[0], model.coef_[1]))

print("MAE: {}".format(np.abs(dataframe['dynamic_time']-predictions).mean()))
print("RMSE: {}".format(np.sqrt(((dataframe['dynamic_time']-predictions)**2).mean())))

coefs = model.coef_
intercept = model.intercept_
xs = np.tile(np.arange(1000000, 5000000, 8000), (500, 1))
ys = np.tile(np.arange(1, 6000, 12), (500, 1)).T
zs = xs.dot(coefs[0])+ys.dot(coefs[1])+intercept

axes.plot_surface(xs,ys,zs, alpha=0.9, color='purple')
axes.plot_trisurf(dataframe['dbsize'], dataframe['query_size'], dataframe['dynamic_time'], alpha=0.2, color='green')

dataframe.set_index(['dbsize', 'query_size', 'dynamic_time'], inplace=True)
dataframe.style.to_latex(
    'test.tex',
    position_float="centering",
    position="htb!",
    hrules=True)
```
